## Intelligent Agents
We call AI programs, intelligent agents. And they sort of have a cycle called perception action cycle. Agent -> sensors -> environment -> actuators -> agent. So the agent has sensors, and the sensors sense stuff from the environment. Because of that sense of environment, the actuators start to act on what the environment is reflecting, and the we give the results back to the agent. So it's a cycle, a loop that runs forever. ## Terminology
We have a lot of things to take in over here. But I need to get better on just one.
* Fully vs Partially Observable
    * Whether sensing the environment is full or partial.
    * If it's full, you can get every information about the environment.
    * If it's partial, you get partial information about the environment.
* Deterministive vs Stochastic
    * Whether the sensed environment commands occur 100% or under 100%.
    * If it's deterministic, commands occur 100%.
    * If it's stochastic, commands occur in a certain probability.
* Continuous vs Discrete
    * Whether the sensed environment is restricted or infinite.
    * If it's continuous, the environment is infinute to sense.
    * If it's discrete, the environment is restricted to sense.
* Adverserial vs Benign
    * Whether the sensed environment has objectives or no objectives.
    * If it's adversrial, the environment gives sensable objectives.
    * If it's benign, the environment doesn't give sensable objectives.

## AI and Uncertainty
So AI is used to manage uncertainty. Yes. We can't really generalize everything. But that's what we want. But the reason we're making generalizations is because we want a 'certain' output from a 'certain' input. It's just like pure functions in functional programming. Everyone wants the same output from the same input. If we put in some input and some other kind of output comes out, that's when we started to get screwed, and hard to rely on. We need things that are reliable. 
 AI makes this reliability. It manages the uncertainty that we can't determine all the time. We get uncertainty from sensor limits, laziness(when we just ignore stuff), ignorance, stochastic environments, adverserials.
